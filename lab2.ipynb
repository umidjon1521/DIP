```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230400/Documents/11.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

red_added = image_RGB.copy()
red_added[:, :, 0] = cv2.add(red_added[:, :, 0], 100)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(red_added)
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('/Users/230400/Documents/11.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

factors = [0.5, 2.0]

plt.figure(figsize=(15, 5))

plt.subplot(1, len(factors) + 1, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

for i, factor in enumerate(factors):
    
    multiplied = cv2.multiply(image_RGB.astype(np.float32), factor)

    multiplied = np.clip(multiplied, 0, 255).astype(np.uint8)
    
    plt.subplot(1, len(factors) + 1, i + 2)
    plt.imshow(multiplied)
    plt.title(f'Multiplied by {factor}')
    plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230400/Documents/11.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

multiplied_image = cv2.divide(image_RGB, 2)

plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multiplied_image)
plt.title('Divided Image 2')
plt.axis('off')
plt.show()
```
